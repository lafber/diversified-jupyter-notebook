# BW Support Package notes number analysis

```python
import csv # ressources data in CSV 
```

```python
# Importation du fichier des données sur le nombre de notes par SP pour le 7.31

my_sp = []
my_nb = []
my_data = {} #dictionnaire pour les données sources

with open('ressources/BW740SP.csv','r') as source_file:
    csv_file = csv.reader(source_file)
    next(csv_file)
    for row in csv_file:
        my_sp.append('SP' + row[0])
        my_nb.append(int(row[3]))
        
my_data['SP'] = my_sp
my_data['nb'] = my_nb
        
print (my_data)
```

```python
# loading Bokeh
from bokeh.charts import Bar, output_notebook, show

# output to notebook
output_notebook() 

# prepare some data
plot = Bar(my_data, label='SP', values='nb', color="#01A1DD", legend=False)
```

```python
show(plot)
```
